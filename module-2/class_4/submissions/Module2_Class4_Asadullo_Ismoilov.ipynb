## Section 3 — Filtering, GroupBy, Merging

### Task 3.1: Selection

### Task 3.2: Boolean Filtering

```python
# Filter 1: First-class passengers older than 50
filter_1 = titanic_df[(titanic_df["Pclass"] == 1) & (titanic_df["Age"] > 50)]

# Filter 2: Female passengers OR passengers paying more than $100
filter_2 = titanic_df[(titanic_df["Sex"] == "female") | (titanic_df["Fare"] > 100)]

display(filter_1.head(3))
```

**Why `&` and Parentheses?**
We must use the bitwise `&` (and `|`) operators in pandas rather than the Python `and` (or `or`) keyword because `and` implicitly attempts to evaluate the *entire* series array's boolean truth value (which is ambiguous), while `&` applies the operation element-wise across the vectorized sequence. Additionally, parentheses `()` are strictly required around conditions because bitwise operators hold higher precedence in Python than comparison operators (like `==` or `>`); without parentheses, it would attempt to evaluate the logic in the wrong order.

### Task 3.3: GroupBy: split-apply-combine

```python
# Pattern 1
mean_age_by_class = titanic_df.groupby("Pclass")["Age"].mean()
print("Mean age by Pclass:")
print(mean_age_by_class)
print("\nReturn type:", type(mean_age_by_class))

# Pattern 2
agg_stats = titanic_df.groupby("Pclass").agg({
    "Age": "mean",
    "PassengerId": "count"
})
print("\nAggregated Stats:")
display(agg_stats)
```

**What type does the first call return?**
The first call (`df.groupby("some_category")["some_numeric"].mean()`) returns a **pandas Series**, with the grouped category (`Pclass`) acting as the underlying index.

### Task 3.4: Merging

```python
# Build a small lookup DataFrame manually
lookup_df = pd.DataFrame({
    "Pclass": [1, 2, 3],
    "Class_Name": ["First Class", "Second Class", "Third Class"],
    "Dining_Location": ["Upper Deck", "Main Deck", "Lower Deck"]
})

# Merge into main DataFrame
merged_df = pd.merge(titanic_df, lookup_df, on="Pclass", how="left")
display(merged_df[["Name", "Pclass", "Class_Name", "Dining_Location"]].head())
```

**What happens to left table rows with no match using "left" join?**
When using a "left" join, any rows in the main (left) DataFrame that fail to find a matching key inside the lookup (right) DataFrame will still be retained in the final output. However, all the resulting columns that were pulled from the right DataFrame for that row will just be populated with `NaN` (null) values since there was no joined data to provide.
