```python
import pandas as pd

df = pd.DataFrame({
    "name": ["Ali", "Zara", "Jasur"],
    "city": ["Tashkent", "Samarqand", "Bukhara"]
})

lookup_df = pd.DataFrame({
    "city": ["Tashkent", "Samarqand"],
    "population": [2500000, 600000]
})

merged_df = pd.merge(df, lookup_df, on="city", how="left")
print(merged_df)
```

Rows in the left table are always kept in the result. If there is no matching value in the lookup table, the new columns will contain NaN (missing values).
