```python
# --- Variables and Data Types ---
# Create variables of different types and print their values and types

age = 18  # integer value (age)
gpa = 3.85  # float value (GPA)
name = "Shakhnoza"  # string value (name)
is_student = True  # boolean value (student status)
scores = [85, 92, 78, 95, 88]  # list of scores

# Print values and their data types
print(age, type(age))
print(gpa, type(gpa))
print(name, type(name))
print(is_student, type(is_student))
print(scores, type(scores))
```

```python
# --- Function to calculate average ---
# This function takes a list of numbers and returns their average

def calculate_average(numbers):
    total = sum(numbers)  # calculate sum
    count = len(numbers)  # count elements
    return total / count  # return average

# Test the function with two different lists
test1 = [80, 90, 70, 85, 95]
test2 = [55, 62, 48, 71]

print("Average 1:", calculate_average(test1))
print("Average 2:", calculate_average(test2))
```

```python
# --- Loop with enumerate ---
# Loop through a list and print both index and value

cities = ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana"]

for index, city in enumerate(cities):
    print(f"Index {index}: {city}")
```

```python
# --- Combine Everything ---
# Create exam scores, calculate average, and check pass/fail

# Create a list of exam scores
exam_scores = [75, 82, 68, 90, 55, 77, 84, 61, 73, 88]

# Print scores
print("Scores:", exam_scores)

# Calculate average
avg = calculate_average(exam_scores)

# Print average score
print(f"Class average: {avg:.1f}")

# Check pass/fail condition
if avg >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
```
