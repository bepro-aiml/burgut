```python
# -----------------************** Task 2: Variables and Data Types **************-----------------
# • Step 1: In a new code cell, create your variables. Example:
age = 22
gpa = 3.85
name = "Aziz"
is_student = True
scores = [85, 92, 78, 95, 88]
# • Step 2: Print each variable with its type:
print(age, type(age))
print(gpa, type(gpa))
print(name, type(name))
print(is_student, type(is_student))
print(scores, type(scores))
# • Step 3: Run the cell (Shift+Enter) and verify the output shows <class 'int'>, <class'float'>, etc.
```

```python
# -----------------************** Task 3: Write a Function **************-----------------
# • Step 1: In a new code cell, define the function:
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count
# • Step 2: Test it with two different lists:
test1 = [80, 90, 70, 85, 95]
test2 = [55, 62, 48, 71]
print("Average 1:", calculate_average(test1))
print("Average 2:", calculate_average(test2))
# • Step 3: Verify the output makes sense (mentally check one of the averages).
```

```python
# -----------------************** Task 4: Loops with enumerate() **************-----------------
# • Step 1: Create a list of at least 5 items (e.g., city names, student names, or course topics).
# • Step 2: Use enumerate() to get both the index and the value:
cities = ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana"]
for index, city in enumerate(cities):
    print(f"Index {index}: {city}")
# • Step 3: Run the cell and verify the output shows indices 0-4 with corresponding items.
```

```python
# -----------------************** Task 5: Combine Everything **************-----------------
import random
exam_scores = [random.randint(0, 100) for _ in range(10)]
print("Scores:", exam_scores)
# • Step 2: Use your calculate_average function:
avg = calculate_average(exam_scores)
print(f"Class average: {avg:.1f}")
# • Step 3: Add a pass/fail check:
if avg >= 60:
  print("Result: PASS")
else:
  print("Result: FAIL")
```
