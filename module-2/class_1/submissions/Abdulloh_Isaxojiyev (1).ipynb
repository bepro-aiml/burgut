## 1. Variables and Data Types

Python has four core data types you will use constantly:
- **int** -- whole numbers (subscriber count, age)
- **float** -- decimal numbers (bill amount, accuracy)
- **str** -- text (customer name, region)
- **bool** -- True or False (is_active, payment_on_time)
- **list** -- ordered collection (monthly bills, feature names)

```python
# --- EXAMPLES ---

# Integers (butun sonlar)
subscriber_count = 10_000_000  # 10 million obunachilar soni (_ faqat o‘qishni osonlashtiradi)
age = 25  # yosh

# Floats (o‘nlik sonlar)
monthly_bill = 89_500.50  # oylik to‘lov (kasr son)
accuracy = 0.95  # aniqlik (95%)

# Strings (matnlar)
customer_name = "Abdullayev Jasur"  # mijoz ismi
region = "Tashkent"  # hudud nomi

# Booleans (True yoki False qiymatlar)
is_active = True  # foydalanuvchi aktivmi (ha)
payment_on_time = False  # to‘lov vaqtida qilinganmi (yo‘q)

# Lists (ro‘yxatlar — bir nechta qiymatni saqlaydi)
monthly_bills = [85000, 92000, 78000, 95000, 88000]  # oylik to‘lovlar ro‘yxati
regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana"]  # shaharlar ro‘yxati

# Check types (har bir o‘zgaruvchining turi va qiymatini chiqaradi)
print(f"subscriber_count: {subscriber_count} (type: {type(subscriber_count).__name__})")
# subscriber_count qiymatini va uning turi (int) ni chiqaradi

print(f"monthly_bill:     {monthly_bill} (type: {type(monthly_bill).__name__})")
# monthly_bill qiymati va turi (float)

print(f"customer_name:    {customer_name} (type: {type(customer_name).__name__})")
# customer_name qiymati va turi (str)

print(f"is_active:        {is_active} (type: {type(is_active).__name__})")
# is_active qiymati va turi (bool)

print(f"monthly_bills:    {monthly_bills} (type: {type(monthly_bills).__name__})")
# monthly_bills ro‘yxati va turi (list)
```

```python
# --- List operations ---

# Access by index (0-based)
print(f"First bill: {monthly_bills[0]}")
# ro‘yxatdagi 1-elementni oladi (index 0 dan boshlanadi)

print(f"Last bill:  {monthly_bills[-1]}")
# ro‘yxatdagi oxirgi elementni oladi (-1 doim oxiri)

# Slicing (kesib olish)
print(f"First 3 bills: {monthly_bills[:3]}")
# 0,1,2 indexlarni oladi (3 kirmaydi) → birinchi 3 ta element

# Length (uzunligi)
print(f"Number of bills: {len(monthly_bills)}")
# ro‘yxatda nechta element borligini hisoblaydi

# Append (yangi element qo‘shish)
monthly_bills.append(91000)
# ro‘yxat oxiriga 91000 ni qo‘shadi

print(f"After append: {monthly_bills}")
# yangilangan ro‘yxatni chiqaradi
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
# --- EXAMPLE: calculate_average ---

def calculate_average(numbers):
    """Calculate the average of a list of numbers."""
    # bu function ro‘yxatdagi sonlarning o‘rtachasini hisoblaydi

    total = sum(numbers)
    # barcha sonlarni qo‘shadi (masalan: 85000 + 92000 + ...)

    count = len(numbers)
    # nechta element borligini hisoblaydi

    return total / count
    # o‘rtacha qiymat = jami / soni

# ro‘yxat (oylik to‘lovlar)
bills = [85000, 92000, 78000, 95000, 88000, 91000]

avg_bill = calculate_average(bills)
# function chaqirildi va natija avg_bill ga saqlandi

print(f"Average monthly bill: {avg_bill:,.0f} soum")
# natijani chiqaradi
# :,.0f → sonni format qiladi:
# ,  → minglik ajratadi (100000 → 100,000)
# .0f → kasr qismini olib tashlaydi (butun son qiladi)
```

```python
# --- EXAMPLE: function with multiple returns ---

def describe_bills(bills):
    """Return basic statistics for a list of bills."""
    # bu function ro‘yxat bo‘yicha o‘rtacha, minimum va maximum qiymatlarni qaytaradi

    avg = sum(bills) / len(bills)
    # o‘rtacha qiymat = jami / elementlar soni

    minimum = min(bills)
    # ro‘yxatdagi eng kichik qiymatni topadi

    maximum = max(bills)
    # ro‘yxatdagi eng katta qiymatni topadi

    return avg, minimum, maximum
    # 3 ta qiymatni bir vaqtning o‘zida qaytaradi (tuple ko‘rinishida)

# function chaqirilmoqda
avg, low, high = describe_bills(bills)
# qaytgan 3 ta qiymat alohida o‘zgaruvchilarga joylanadi:
# avg → o‘rtacha
# low → minimum
# high → maximum

print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")
# natijalarni chiqaradi
# :,.0f → sonni formatlaydi (minglik ajratadi va kasrni olib tashlaydi)
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
# --- EXAMPLE: basic for loop ---

regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]
# bu ro‘yxat (list) — ichida bir nechta shahar nomlari bor

for region in regions:
    # for loop → ro‘yxatdagi har bir elementni bittadan olib chiqadi
    # region → har aylanishda (iteration) navbatdagi qiymatni oladi

    print(f"Processing data for: {region}")
    # har bir shahar uchun matn chiqaradi
    # masalan: Processing data for: Tashkent
```

```python
# --- EXAMPLE: enumerate (index + value) ---
# enumerate → index (pozitsiya) + value (qiymat) ni birga beradi

monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]
# oylik to‘lovlar ro‘yxati

for i, bill in enumerate(monthly_bills):
    # enumerate → har bir element uchun:
    # i → index (0 dan boshlanadi)
    # bill → o‘sha indexdagi qiymat

    print(f"Month {i + 1}: {bill:,} soum")
    # i + 1 → 0 emas, 1-dan boshlanishi uchun
    # {bill:,} → sonni format qiladi (85000 → 85,000)
```

```python
# --- EXAMPLE: loop with conditional ---

customers = [
    {"name": "Alisher", "spend": 200000},   # 1-mijoz
    {"name": "Dilnoza", "spend": 45000},    # 2-mijoz
    {"name": "Bobur",   "spend": 120000},   # 3-mijoz
    {"name": "Malika",  "spend": 35000},    # 4-mijoz
    {"name": "Rustam",  "spend": 180000},   # 5-mijoz
]
# bu list ichida dictionary lar bor (har biri bitta mijoz)

print("High-value customers (spend > 100,000):")
# sarlavha chiqaradi

for c in customers:
    # for loop → har bir mijozni bittadan oladi
    # c → bitta mijoz (dictionary)

    if c["spend"] > 100000:
        # agar mijozning xarajati 100000 dan katta bo‘lsa

        print(f"  {c['name']}: {c['spend']:,} soum")
        # mijoz ismi va xarajatini chiqaradi
        # {c['spend']:,} → sonni 200000 → 200,000 ko‘rinishda chiqaradi
```

## 5. Putting It All Together

Combine variables, functions, and loops to solve a mini data task.

```python
customers_data = [
    {"id": 1, "name": "Anvar",   "monthly_spend": 210000},
    {"id": 2, "name": "Nilufar", "monthly_spend": 55000},
    {"id": 3, "name": "Sardor",  "monthly_spend": 180000},
    {"id": 4, "name": "Gulnora", "monthly_spend": 32000},
    {"id": 5, "name": "Timur",   "monthly_spend": 95000},
    {"id": 6, "name": "Madina",  "monthly_spend": 160000},
    {"id": 7, "name": "Otabek",  "monthly_spend": 47000},
    {"id": 8, "name": "Zarina",  "monthly_spend": 125000},
]

# Har bir kategoriya uchun hisoblagich (counter) ochyapmiz
premium = 0   # premium mijozlar soni
standard = 0  # standard mijozlar soni
basic = 0     # basic mijozlar soni

# Har bir customer ustidan aylanish (loop)
for customer in customers_data:

    # Har bir customerning oylik xarajatini olish
    spend = customer["monthly_spend"]

    # Agar 150,000 yoki undan ko'p sarflasa → Premium
    if spend >= 150000:
        premium += 1  # premium sonini 1 ga oshiramiz

    # Agar 80,000 dan 149,999 gacha bo‘lsa → Standard
    elif spend >= 80000:
        standard += 1  # standard sonini oshiramiz

    # Qolgan hammasi (< 80,000) → Basic
    else:
        basic += 1  # basic sonini oshiramiz

# Natijani chiroyli chiqarish
print(f"Premium: {premium}, Standard: {standard}, Basic: {basic}")
```
