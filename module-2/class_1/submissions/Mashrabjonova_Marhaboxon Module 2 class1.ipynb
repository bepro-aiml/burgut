## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
# --- EXAMPLE: basic for loop ---

# regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]

# for region in regions:
#     print(f"Processing data for: {region}")
for loop: Used to iterate (repeat) over a sequence like a list.
Variable region: In each step, this variable holds the current item from the list.
Automation: Instead of printing 5 times manually, the loop does it automatically for every item in the list.
```

```python
# --- EXAMPLE: enumerate (index + value) ---
# enumerate gives you both the position and the value

# monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

# for i, bill in enumerate(monthly_bills):
#     print(f"Month {i + 1}: {bill:,} soum")
enumerate(): Allows you to loop through a list and get both the index (i) and the value (bill) at the same time.
Purpose: It's useful when you need to know the position of an item (e.g., "Month 1", "Month 2").
i + 1: Since Python starts counting at 0, we add 1 to make it human-readable.
```
