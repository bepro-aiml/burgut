## 2. Dictionaries

A dictionary stores key-value pairs. Think of it as a single row in a database table -- keys are column names, values are the data.

```python
# --- EXAMPLE ---

customer = {
    "id": "UZT-001234",
    "name": "Karimov Sherzod",
    "region": "Tashkent",
    "monthly_spend": 125000,
    "is_active": True,
    "plan": "Premium"
}

# Access values by key
print(f"Customer: {customer['name']}")
print(f"Region:   {customer['region']}")
print(f"Spend:    {customer['monthly_spend']} soum")

# Add a new key
customer["tenure_months"] = 24
print(f"\nFull record: {customer}")
```

```python
# --- Explanation of Dictionary Example ---

# A dictionary in Python stores data as key-value pairs.
# It is similar to a row in a database, where:
# keys are like column names, and values are the actual data.

# In this example, we create a dictionary called "customer"
# that contains information about a telecom client.

# Each key represents a field:
# "id" -> customer ID
# "name" -> customer name
# "region" -> location
# "monthly_spend" -> amount of money spent
# "is_active" -> whether the customer is active (True/False)
# "plan" -> subscription plan type

# Accessing values:
# We use customer['key'] to get specific values from the dictionary.

# For example:
# customer['name'] returns the customer's name
# customer['region'] returns the region

# Adding a new key:
# We can add new data by assigning a value to a new key
# customer["tenure_months"] = 24 adds how long the customer has been subscribed

# This example shows how dictionaries are used to store and manage structured data
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
# --- EXAMPLE: calculate_average ---

def calculate_average(numbers):
    """Calculate the average of a list of numbers."""
    total = sum(numbers)
    count = len(numbers)
    return total / count

bills = [85000, 92000, 78000, 95000, 88000, 91000]
avg_bill = calculate_average(bills)
print(f"Average monthly bill: {avg_bill:,.0f} soum")
```

```python
# --- Explanation of Function Example ---

# A function is a reusable block of code that performs a specific task.
# It takes input (parameters), processes it, and returns an output.

# In this example, we define a function called calculate_average.

# "def" is used to define a function.
# "numbers" is the input parameter (a list of numbers).

# Inside the function:
# sum(numbers) calculates the total sum of all elements in the list.
# len(numbers) returns the number of elements in the list.

# Then we calculate the average by dividing total by count.

# "return" sends the result back to where the function was called.

# After defining the function, we create a list called "bills".

# We call the function using calculate_average(bills)
# and store the result in avg_bill.

# Finally, we print the average monthly bill.

# This demonstrates how functions help organize code and reuse logic.
```

```python



# --- EXAMPLE: function with multiple returns ---

def describe_bills(bills):
    """Return basic statistics for a list of bills."""
    avg = sum(bills) / len(bills)
    minimum = min(bills)
    maximum = max(bills)
    return avg, minimum, maximum

avg, low, high = describe_bills(bills)
print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")
```

```python
# --- Explanation of Function with Multiple Returns ---

# This function describes basic statistics for a list of bills.

# The function name is "describe_bills" and it takes one input: bills (a list of numbers).

# Inside the function:
# We calculate the average by dividing sum(bills) by len(bills).
# min(bills) returns the smallest value in the list.
# max(bills) returns the largest value in the list.

# The function returns three values: average, minimum, and maximum.

# When calling the function:
# avg, low, high = describe_bills(bills)

# This means:
# avg receives the average value
# low receives the minimum value
# high receives the maximum value

# Finally, we print all three values in a formatted string.

# This example shows that a function can return multiple results at once.
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python

# --- EXAMPLE: basic for loop ---

regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]

for region in regions:
    print(f"Processing data for: {region}")
```

```python
# --- Explanation of Basic For Loop ---

# A "for" loop is used to iterate over a sequence (like a list).

# In this example, "regions" is a list of city names.

# The loop goes through each element in the list one by one.

# "region" is a temporary variable that takes each value from the list.

# For each region, the program prints a message.

# This means the print statement runs multiple times,
# once for each item in the list.

# This example shows how loops help automate repetitive tasks.
```

```python



# --- EXAMPLE: enumerate (index + value) ---
# enumerate gives you both the position and the value

monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

for i, bill in enumerate(monthly_bills):
    print(f"Month {i + 1}: {bill:,} soum")
```

```python
# --- Explanation of enumerate() in a for loop ---

# enumerate() is used when we need both the index (position)
# and the value from a list at the same time.

# In this example, "monthly_bills" is a list of bill amounts.

# The loop goes through each element in the list.

# "i" represents the index (starting from 0)
# "bill" represents the actual value from the list

# We use (i + 1) because index starts at 0,
# but we want to display months starting from 1

# The print statement shows the month number and bill amount

# The format {bill:,} adds commas for better readability

# This example shows how enumerate() helps track position and value together
```

```python



























# --- EXAMPLE: loop with conditional ---

customers = [
    {"name": "Alisher", "spend": 200000},
    {"name": "Dilnoza", "spend": 45000},
    {"name": "Bobur",   "spend": 120000},
    {"name": "Malika",  "spend": 35000},
    {"name": "Rustam",  "spend": 180000},
]

print("High-value customers (spend > 100,000):")
for c in customers:
    if c["spend"] > 100000:
        print(f"  {c['name']}: {c['spend']:,} soum")
```

```python
# --- Explanation of Loop with Conditional ---

# This example combines a loop with a conditional statement (if).

# "customers" is a list of dictionaries.
# Each dictionary represents one customer with a name and spend value.

# The loop goes through each customer in the list.

# "c" is a temporary variable representing each customer.

# Inside the loop, we check a condition:
# if the customer's spend is greater than 100000

# c["spend"] accesses the spend value from the dictionary

# If the condition is True, we print the customer's name and spend

# This filters only high-value customers

# The format {c['spend']:,} is used to display numbers with commas

# This example shows how loops and conditions work together
# to filter and process specific data
```
