## 1. Variables and Data Types

Python has four core data types you will use constantly:
- **int** -- whole numbers (subscriber count, age)
- **float** -- decimal numbers (bill amount, accuracy)
- **str** -- text (customer name, region)
- **bool** -- True or False (is_active, payment_on_time)
- **list** -- ordered collection (monthly bills, feature names)

## 2. Dictionaries

A dictionary stores key-value pairs. Think of it as a single row in a database table -- keys are column names, values are the data.

```python
#-----------------************** FOURTH CELL **************-----------------
# --- EXAMPLE ---

customer = {
    "id": "UZT-001234",
    "name": "Karimov Sherzod",
    "region": "Tashkent",
    "monthly_spend": 125000,
    "is_active": True,
    "plan": "Premium"
}

# Access values by key
print(f"Customer: {customer['name']}")
print(f"Region:   {customer['region']}")
print(f"Spend:    {customer['monthly_spend']} soum")

# Add a new key
customer["tenure_months"] = 24
print(f"\nFull record: {customer}")
```

```python
# FOURTH CELL explanation:
# 1. First declaring Variable "customer" with Dictionary Method(Key: "id", "name".....    Value: "UZT-001234", "Karimov Sherzod" ....)

# 2. Printing Customer, Region, Spend with accessing values by their keys respectively

# 3. Adding a new key with its value ("tenure_months": 24) to the existing dictionary
```

## 3. Functions

Functions take inputs, perform an operation, and return an output. They let you write reusable, modular code.

In ML, every step of a pipeline is typically a function: load_data(), clean_data(), train_model(), evaluate().

```python
#-----------------************** FIFTH CELL **************-----------------
# --- EXAMPLE: calculate_average ---

def calculate_average(numbers):
    """Calculate the average of a list of numbers."""
    total = sum(numbers)
    count = len(numbers)
    return total / count

bills = [85000, 92000, 78000, 95000, 88000, 91000]
avg_bill = calculate_average(bills)
print(f"Average monthly bill: {avg_bill:,.0f} soum")
```

```python
# FIFTH CELL explanation:
# 1. Creating Function (calculate_average) starts with a description """Calculate the average of a list of numbers."""
#    And declaring "total" and "count" which finds sum of numbers and amount of numbers respectively
#    Returning "total" divided by "count" for calculating average

# 2. List (array) named "bills" created and entered integer values

# 3. Called Function calculate_average for calculating average of variable bills. And Printing average value of list
```

```python
#-----------------************** SIXTH CELL **************-----------------
# --- EXAMPLE: function with multiple returns ---

def describe_bills(bills):
    """Return basic statistics for a list of bills."""
    avg = sum(bills) / len(bills)
    minimum = min(bills)
    maximum = max(bills)
    return avg, minimum, maximum

avg, low, high = describe_bills(bills)
print(f"Average: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")
```

```python
# SIXTH CELL explanation:
# 1. Creating Function (describe_bills) which takes variable "bills" that was entered FIFTH CELL. And Function starts with a description """Return basic statistics for a list of bills."""
#    And declaring "avg", "minimum" and "maximum" which finds average (sum divided by length), minimum and maximum of bills respectively
#    Returning Average, Minimum and Maximum.

# 2. Called Function describe_bills for calculating average of numbers, minimum and maximum number in variable "bills". And Printing Them into screen. (.0f is for rounding up)
```

## 4. Loops

Loops let you repeat an operation across a collection. In data work, you iterate over rows, columns, files, or parameter values constantly.

```python
#-----------------************** SEVENTH CELL **************-----------------
# --- EXAMPLE: basic for loop ---

regions = ["Tashkent", "Samarkand", "Bukhara", "Fergana", "Namangan"]

for region in regions:
    print(f"Processing data for: {region}")
```

```python
# SEVENTH CELL explanation:
# 1. Creating List named "regions" contains string elements.

# 2. Using "for" loop for printing every element of regions one by one with description "Processing data for: "
```

```python
#-----------------************** EIGHTH CELL **************-----------------
# --- EXAMPLE: enumerate (index + value) ---
# enumerate gives you both the position and the value

monthly_bills = [85000, 92000, 78000, 95000, 88000, 91000]

for i, bill in enumerate(monthly_bills):
    print(f"Month {i + 1}: {bill:,} soum")
```

```python
# EIGHTH CELL explanation:
# 1. Creating List named "monthly_bills" contains integer elements.

# 2. "For" loop created with variable "i" inside and variable 'i' represents the current index, and 'bill' represents the current value

# 3. And Printing next element with adding 1 to index
```

```python
#-----------------************** NINETH CELL **************-----------------
# --- EXAMPLE: loop with conditional ---

customers = [
    {"name": "Alisher", "spend": 200000},
    {"name": "Dilnoza", "spend": 45000},
    {"name": "Bobur",   "spend": 120000},
    {"name": "Malika",  "spend": 35000},
    {"name": "Rustam",  "spend": 180000},
]

print("High-value customers (spend > 100,000):")
for c in customers:
    if c["spend"] > 100000:
        print(f"  {c['name']}: {c['spend']:,} soum")
```

```python
# NINETH CELL explanation:
# 1. First declaring Variable "customers" with List of Dictionaries Method(Keys: "spend", "name", .....    Values: "200,000", "Alisher" ....)

# 2. Printing with If conditional to filter customers. It says PRINT when spend value is larger than 100,000, and Print name too
```

```python
# TENTH CELL explanation:
# 1. Creating a list of 10 random exam scores.
# 2. Calculating the average using our previously defined function.
# 3. Checking if the average is 60 or higher to determine pass/fail status.

exam_scores = [85, 45, 90, 55, 70, 60, 30, 80, 95, 40] # 10 ta tasodifiy ball
avg = calculate_average(exam_scores) # Funksiyani chaqirish

print(f"Scores: {exam_scores}")
print(f"Class average: {avg:.1f}")

if avg >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
```
