# **Mod_2 Assignment 2**

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("vivek468/superstore-dataset-final")

print("Path to dataset files:", path)
```

```python
# -----Task 2: Inspect the Structure-------
import pandas as pd

# Load the dataset from a CSV file into a Pandas DataFrame
df = pd.read_csv('/kaggle/input/superstore-dataset-final/Sample - Superstore.csv', encoding='latin1')

# Display the first 10 rows to verify the data was loaded correctly
df.head(10)
```

```python
df.info()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
# -----Task 3: Handle Missing Values-------
# Fill missing numerical values with the median
df['Sales'] = df['Sales'].fillna(df['Sales'].median())

# Fill missing categorical (text) values with the mode (most common value)
df['Ship Mode'] = df['Ship Mode'].fillna(df['Ship Mode'].mode()[0])

# Verify that there are no more missing values
print(df.isnull().sum())
```

```python
# ------Task 4: Remove Duplicates ------
# Check the number of duplicate rows
print(f"Duplicates found: {df.duplicated().sum()}")

# Remove duplicate rows
df = df.drop_duplicates()

# Verify the new shape of the dataset
print(f"New shape after removing duplicates: {df.shape}")
```

```python
# -----Task 5: Convert Data Types-------
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
print(df.dtypes)
df[['Order Date', 'Ship Date']].head()
```

```python
# -----Task 6: Create Derived Features-------
# Step 1: Calculate total spending per customer
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spent']

# Step 2: Calculate order frequency (how many times they shopped)
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_count']

# Step 3: Merge these into a customer summary table
customer_summary = total_spending.merge(order_freq, on='Customer ID')

# Step 4: Show the first 10 rows of the summary
customer_summary.head(10)
```

```python
# Save the cleaned data to a new CSV file
df.to_csv('superstore_cleaned.csv', index=False)

# Save the customer summary to a new CSV file
customer_summary.to_csv('customer_summary.csv', index=False)

print("Files saved successfully!")
```

# **Mod_2 Assignment 3**

```python
# Setup
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('superstore_cleaned.csv')
df['Order Date'] = pd.to_datetime(df['Order Date'])

df.head()
```

```python
# Task 1
plt.figure(figsize=(10, 6))
plt.hist(df['Sales'], bins=50, color='steelblue', edgecolor='black')
plt.title('Distribution of Sales')
plt.xlabel('Sales ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Profit'], color='lightcoral')
plt.title('Boxplot of Profit')
plt.xlabel('Profit ($)')
plt.show()
```

```python
# Task 2
cat_sales = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
cat_sales.plot(kind='bar', color=['#2196F3', '#4CAF50', '#FF9800'])
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.show()
```

```python
region_sales = df.groupby('Region')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
region_sales.plot(kind='bar', color='steelblue')
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=0)
plt.show()
```

```python
# Task 3
Q1 = df['Profit'].quantile(0.25)
Q3 = df['Profit'].quantile(0.75)
IQR = Q3 - Q1
print(f"Q1: {Q1:.2f}, Q3: {Q3:.2f}, IQR: {IQR:.2f}")

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(f"Lower bound: {lower_bound:.2f}, Upper bound: {upper_bound:.2f}")

outliers = df[(df['Profit'] < lower_bound) | (df['Profit'] > upper_bound)]
print(f"Number of outliers: {len(outliers)} out of {len(df)} rows ({len(outliers)/len(df)*100:.1f}%)")
```

```python
# Task 4
numerical_cols = df[['Sales', 'Quantity', 'Discount', 'Profit']]
corr_matrix = numerical_cols.corr()
print(corr_matrix)

plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Heatmap: Sales, Quantity, Discount, Profit')
plt.show()
```

```python
# Task 5
monthly_sales = df.resample('M', on='Order Date')['Sales'].sum()

plt.figure(figsize=(12, 6))
monthly_sales.plot(color='steelblue', linewidth=2)
plt.title('Monthly Total Sales Over Time')
plt.xlabel('Month')
plt.ylabel('Total Sales ($)')
plt.grid(True, alpha=0.3)
plt.show()
```

# **Mod_2 Assignment 5**

```python
# Step 0
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
```

```python
# Step 1
counts = df['Contract'].value_counts()
plt.figure(figsize=(7, 4))
plt.bar(counts.index, counts.values, color='teal')
plt.title('Customers per Contract type')
plt.ylabel('Number of customers')
plt.show()
```

```python
# Step 2
plt.figure(figsize=(8, 4))
plt.hist(df['tenure'], bins=30, color='coral', edgecolor='black')
plt.title('Distribution of customer tenure')
plt.xlabel('Tenure (months)')
plt.ylabel('Number of customers')
plt.show()
```

```python
# Step 3
plt.figure(figsize=(8, 5))
plt.scatter(df['MonthlyCharges'], df['TotalCharges'], alpha=0.3, s=10, color='navy')
plt.title('Monthly vs Total Charges')
plt.xlabel('Monthly Charges')
plt.ylabel('Total Charges')
plt.show()
```

```python
# Step 4
groups = [df[df['Contract'] == c]['MonthlyCharges'] for c in df['Contract'].unique()]
labels = df['Contract'].unique()
plt.figure(figsize=(7, 5))
plt.boxplot(groups, labels=labels)
plt.title('Monthly Charges by Contract')
plt.ylabel('Monthly Charges')
plt.show()
```

```python
# Step 5
fig, ax = plt.subplots(1, 3, figsize=(15, 4))
ax[0].bar(df['Churn'].value_counts().index, df['Churn'].value_counts().values, color='gray')
ax[0].set_title('Churn count')
ax[1].hist(df['MonthlyCharges'], bins=30, color='orange')
ax[1].set_title('Monthly Charges')
ax[2].scatter(df['tenure'], df['TotalCharges'], alpha=0.2, s=5)
ax[2].set_title('Tenure vs Total Charges')
plt.tight_layout()
plt.show()
```
