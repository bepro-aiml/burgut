## Section 1 — Exploratory vs Explanatory

### Task 1.1 — Exploratory vs Explanatory (in my own words)

**Exploratory** visualization is what you do *for yourself* during analysis: you are digging through the data, testing ideas, and trying to understand what is there. The charts are quick and rough; they are a thinking tool, not a communication tool.

**Explanatory** visualization is what you build *for an audience* after you already know the story. Every design choice (color, label, axis range) exists to deliver one clear message to the viewer as efficiently as possible.

In this assignment I am mostly producing **exploratory** charts. I am reproducing lecture templates on a familiar dataset, not crafting a polished narrative for a specific audience. Honest answer: exploratory.

### Task 1.2 — Anscombe's Quartet

The lecture mentions Anscombe's quartet to prove that four completely different datasets can share identical summary statistics (mean, variance, correlation) yet look nothing alike when plotted, which is why you should always visualize your data rather than trusting numbers alone.

## Section 2 — Matplotlib and Seaborn

### Task 2.1 — Plain Matplotlib

```python
# Histogram of passenger age
plt.figure(figsize=(8, 5))
plt.hist(df['age'].dropna(), bins=20, color='steelblue', edgecolor='white')
plt.xlabel('Age (years)')
plt.ylabel('Number of passengers')
plt.title('Distribution of Passenger Ages - Titanic')
plt.tight_layout()
plt.show()
```

```python
# Scatter: age vs fare
plt.figure(figsize=(8, 5))
plt.scatter(df['age'], df['fare'], alpha=0.4, color='coral', edgecolors='none')
plt.xlabel('Age (years)')
plt.ylabel('Fare (GBP)')
plt.title('Age vs Fare - Titanic')
plt.tight_layout()
plt.show()
```

### Task 2.2 — Seaborn (preferred for statistical work)

```python
# KDE-enhanced histogram of age
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(df, x='age', kde=True, bins=20, color='steelblue', ax=ax)
ax.set_xlabel('Age (years)')
ax.set_ylabel('Count')
ax.set_title('Age Distribution with KDE - Titanic')
plt.tight_layout()
plt.show()
```

```python
# Boxplot: fare by passenger class
fig, ax = plt.subplots(figsize=(8, 5))
sns.boxplot(data=df, x='class', y='fare', palette='Set2', ax=ax)
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Fare (GBP)')
ax.set_title('Fare Distribution by Passenger Class - Titanic')
plt.tight_layout()
plt.show()
```

```python
# Scatterplot: age vs fare, coloured by survival
fig, ax = plt.subplots(figsize=(8, 5))
sns.scatterplot(data=df, x='age', y='fare', hue='survived',
                palette={0: 'salmon', 1: 'seagreen'}, alpha=0.6, ax=ax)
ax.set_xlabel('Age (years)')
ax.set_ylabel('Fare (GBP)')
ax.set_title('Age vs Fare coloured by Survival - Titanic')
ax.legend(title='Survived', labels=['No', 'Yes'])
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap of numeric columns
fig, ax = plt.subplots(figsize=(8, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, fmt='.2f',
            cmap='coolwarm', linewidths=0.5, ax=ax)
ax.set_title('Correlation Matrix - Titanic (numeric features)')
plt.tight_layout()
plt.show()
```

### Task 2.3 — Object-Oriented Matplotlib (saved to file)

```python
# OO pattern: survival rate by passenger class
survival_rate = df.groupby('class')['survived'].mean() * 100

fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(survival_rate.index, survival_rate.values,
       color=['#4e79a7', '#f28e2b', '#e15759'], edgecolor='white', width=0.5)
ax.set_xlabel('Passenger Class')
ax.set_ylabel('Survival Rate (%)')
ax.set_title('Survival Rate by Passenger Class - Titanic')
ax.set_ylim(0, 100)
for i, v in enumerate(survival_rate.values):
    ax.text(i, v + 1.5, f'{v:.1f}%', ha='center', fontsize=11, fontweight='bold')
fig.tight_layout()
fig.savefig('out.png', dpi=150)
print('Chart saved to out.png')
plt.show()
```

## Section 3 — Five Rules for Honest Charts

### Task 3.1 — Audit of Task 2.3 chart against the Five Rules

**Chart audited:** Survival Rate by Passenger Class (bar chart, Task 2.3)

| # | Rule | Result | Explanation |
|---|------|--------|-------------|
| 1 | Are axes labelled with units? | **Yes** | X-axis: 'Passenger Class'; Y-axis: 'Survival Rate (%)'. The unit % is explicit. |
| 2 | Does the y-axis start at zero? | **Yes** | `ax.set_ylim(0, 100)` forces the bar chart to start at zero, so bar heights are proportional to real values. |
| 3 | Is it the right chart type? | **Yes** | Comparing one numeric value across a few discrete categories is exactly what a bar chart is for. |
| 4 | Is there any chartjunk? | **No (none present)** | No 3-D effects, no shadows, no gradient fills, no decorative icons. Colors are functional (one per class). |
| 5 | One clear message or many? | **Yes (one message)** | The chart answers exactly one question: which class survived at the highest rate? Data labels reinforce that single takeaway. |

### Task 3.2 — Log Scale Moment

```python
# 'fare' has a heavy right tail: most passengers paid little, a few paid enormous amounts
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Linear y-axis
axes[0].hist(df['fare'].dropna(), bins=40, color='steelblue', edgecolor='white')
axes[0].set_xlabel('Fare (GBP)')
axes[0].set_ylabel('Count')
axes[0].set_title('Fare Distribution - Linear scale')

# Log y-axis
axes[1].hist(df['fare'].dropna(), bins=40, color='coral', edgecolor='white')
axes[1].set_yscale('log')
axes[1].set_xlabel('Fare (GBP)')
axes[1].set_ylabel('Count (log scale)')
axes[1].set_title('Fare Distribution - Log scale')

plt.suptitle('Effect of Log Scale on a Right-Skewed Distribution', fontsize=13)
plt.tight_layout()
plt.show()
```

**What changed and why:**  
On the linear scale the distribution looks like a single spike near zero with an almost invisible tail; switching to a log y-axis compresses the towering low-fare bars and stretches the near-zero high-fare bars upward, revealing that rare fares above 200 GBP still form a meaningful and previously invisible part of the distribution.
