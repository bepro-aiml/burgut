````python
# === SETUP — run this first ===
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings; warnings.filterwarnings('ignore')

url = 'https://raw.githubusercontent.com/IBM/telco-customer-churn-on-icp4d/master/data/Telco-Customer-Churn.csv'
df = pd.read_csv(url)
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce').fillna(0)
print('Loaded:', df.shape)
````

````python
# Cell 0 explanation:
# Import pandas, numpy, matplotlib and suppress warnings
# Load the Telco Customer Churn dataset from GitHub and clean the TotalCharges column
````

### Cell 1 — bar chart of Churn count

````python
counts = df['Churn'].value_counts()
plt.figure(figsize=(6, 4))
plt.bar(counts.index, counts.values, color=['#2E86AB', '#E63946'])
plt.title('Churned vs Stayed')
plt.ylabel('Number of customers')
plt.show()
````

````python
# Cell 1 explanation:
# Count how many customers churned vs stayed, then plot it as a bar chart
````

### Cell 2 — histogram of MonthlyCharges

````python
plt.figure(figsize=(8, 4))
plt.hist(df['MonthlyCharges'], bins=30, color='#2E86AB', edgecolor='black')
plt.title('Distribution of MonthlyCharges')
plt.xlabel('Monthly Charges ($)')
plt.ylabel('Number of customers')
plt.show()
````

````python
# Cell 2 explanation:
# Plot a histogram showing the distribution of MonthlyCharges across all customers
````

### Cell 3 — scatter of tenure vs TotalCharges

````python
plt.figure(figsize=(8, 5))
plt.scatter(df['tenure'], df['TotalCharges'], alpha=0.3, s=10)
plt.title('Tenure vs Total Charges')
plt.xlabel('Tenure (months)')
plt.ylabel('Total Charges ($)')
plt.show()
````

````python
# Cell 3 explanation:
# Plot a scatter chart comparing tenure and TotalCharges to see their relationship
````

### Cell 4 — box plot of tenure by Churn

````python
churn_yes = df[df['Churn'] == 'Yes']['tenure']
churn_no = df[df['Churn'] == 'No']['tenure']
plt.figure(figsize=(6, 5))
plt.boxplot([churn_no, churn_yes], labels=['Stayed', 'Churned'])
plt.title('Tenure distribution by Churn')
plt.ylabel('Tenure (months)')
plt.show()
````

````python
# Cell 4 explanation:
# Split tenure into two groups (Churned vs Stayed) and compare them with a boxplot
````

### Cell 5 — 2x2 subplots dashboard

````python
fig, ax = plt.subplots(2, 2, figsize=(12, 8))
ax[0, 0].bar(df['Churn'].value_counts().index, df['Churn'].value_counts().values, color='steelblue')
ax[0, 0].set_title('Churn')
ax[0, 1].hist(df['MonthlyCharges'], bins=30, color='orange')
ax[0, 1].set_title('MonthlyCharges')
ax[1, 0].scatter(df['tenure'], df['MonthlyCharges'], alpha=0.2, s=5)
ax[1, 0].set_title('Tenure vs MonthlyCharges')
ax[1, 1].hist(df['tenure'], bins=30, color='green')
ax[1, 1].set_title('Tenure')
plt.tight_layout()
plt.show()
````

````python
# Cell 5 explanation:
# Build a 2x2 dashboard of subplots showing Churn, MonthlyCharges, Tenure vs MonthlyCharges, and Tenure distributions together
````
