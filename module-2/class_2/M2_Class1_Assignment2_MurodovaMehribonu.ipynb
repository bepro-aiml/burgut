```python

# DICTIONARY - kalit:qiymat juftliklaridan iborat ma'lumot turi

customer = {
    "id": "UZT-001234",
    "name": "Karimov Sherzod",
    "region": "Tashkent",
    "monthly_spend": 125000,
    "is_active": True,
    "plan": "Premium"
}

# Kalitlar orqali qiymatlarni olish
print(f"Mijoz:    {customer['name']}")
print(f"Hudud:    {customer['region']}")
print(f"Xarajat:  {customer['monthly_spend']} so'm")

# Mavjud dictionary'ga yangi kalit qo'shish
customer["tenure_months"] = 24
print(f"\nTo'liq ma'lumot: {customer}")



# VAZIFA: E-commerce mahsulotini ifodalovchi dictionary yarating
# Kamida 5 ta kalit: name, category, price, quantity_sold, rating


# Mahsulot dictionary'sini yaratamiz
product = {
    "name": "iPhone 15",
    "category": "Smartfonlar",
    "price": 12_000_000,       # narx so'mda
    "quantity_sold": 340,       # sotilgan dona soni
    "rating": 4.8               # foydalanuvchi bahosi (5 dan)
}

# Mahsulot nomi va narxini chiqarish
print(f"Mahsulot: {product['name']}")
print(f"Narxi:    {product['price']} so'm")

# Yangi "discount" kalitini qo'shish (0 dan 1 gacha — masalan 0.15 = 15%)
product["discount"] = 0.15

# .keys() barcha kalitlarni qaytaradi, for sikli bilan chiqaramiz
print("\nBarcha kalitlar:")
for kalit in product.keys():
    print(f"  - {kalit}")


=
# FUNKSIYA — qayta ishlatiladigan kod bloki
# def kalit so'zi bilan e'lon qilinadi


def calculate_average(numbers):
    """Sonlar ro'yxatining o'rtachasini hisoblaydi."""
    total = sum(numbers)    # barcha sonlar yig'indisi
    count = len(numbers)    # sonlar soni
    return total / count    # o'rtacha = yig'indi / son

bills = [85000, 92000, 78000, 95000, 88000, 91000]
avg_bill = calculate_average(bills)
print(f"O'rtacha oylik to'lov: {avg_bill:,.0f} so'm")



# Bir vaqtda bir nechta qiymat qaytaruvchi funksiya
# return orqali vergul bilan ajratib yoziladi


def describe_bills(bills):
    """To'lovlar ro'yxati uchun asosiy statistikani qaytaradi."""
    avg = sum(bills) / len(bills)   # o'rtacha
    minimum = min(bills)             # eng kichik qiymat
    maximum = max(bills)             # eng katta qiymat
    return avg, minimum, maximum     # uchta qiymat birga qaytariladi

# Qaytgan uchta qiymatni alohida o'zgaruvchilarga yozamiz
avg, low, high = describe_bills(bills)
print(f"O'rtacha: {avg:,.0f} | Min: {low:,.0f} | Max: {high:,.0f}")



# VAZIFA 1: classify_customer funksiyasi
# monthly_spend qiymatiga qarab mijoz turini aniqlaydi


def classify_customer(monthly_spend):
    """
    Oylik xarajatga qarab mijozni turkumlaydi:
      - 150000 dan yuqori  → "premium"
      - 50000 dan 150000   → "standard"
      - 50000 dan past     → "basic"
    """
    if monthly_spend > 150000:
        return "premium"
    elif monthly_spend >= 50000:   # 50000 va 150000 oralig'i
        return "standard"
    else:
        return "basic"

# Funksiyani turli qiymatlar bilan sinash
print(classify_customer(200000))   # premium
print(classify_customer(75000))    # standard
print(classify_customer(30000))    # basic



# VAZIFA 2: calculate_churn_rate funksiyasi
# Churn rate = ketgan mijozlar / jami mijozlar * 100


def calculate_churn_rate(total_customers, churned_customers):
    """
    Mijozlar churn foizini hisoblaydi.
    Masalan: 10000 dan 350 tasi ketsa → 3.5%
    """
    churn_rate = (churned_customers / total_customers) * 100
    return churn_rate

# Funksiyani sinash
print(calculate_churn_rate(10000, 350))   # 3.5
```

```python

# VAZIFA 2: enumerate bilan rank va HIGH belgisi chiqarish
=

scores = [72, 95, 88, 61, 93, 77, 85, 91, 68, 99]

for i, score in enumerate(scores):
    rank = i + 1                      # i 0 dan boshlanadi, rank 1 dan boshlanishi kerak

    if score > 90:
        print(f"  Rank {rank}: {score} -- HIGH")   # 90 dan yuqori bo'lsa belgi qo'shamiz
    else:
        print(f"  Rank {rank}: {score}")            # oddiy holda faqat rank va ball



# VAZIFA 3: Jami va o'rtachani QO'LDA hisoblash
# sum() va len() ishlatmasdan, faqat sikl va o'zgaruvchilar bilan


total = 0      # yig'indini saqlovchi o'zgaruvchi, boshida 0
count = 0      # elementlar sonini saqlovchi o'zgaruvchi, boshida 0

for score in scores:
    total += score    # har bir ballni yig'indiga qo'shamiz (total = total + score)
    count += 1        # har bir elementda hisoblagichni 1 ga oshiramiz

average = total / count    # o'rtacha = yig'indi / soni

print(f"Jami: {total} | O'rtacha: {average}")
```

```python

# YAKUNIY MASHQ: Mijozlarni turkumlash va hisobot chiqarish

customers_data = [
    {"id": 1, "name": "Anvar",   "monthly_spend": 210000},
    {"id": 2, "name": "Nilufar", "monthly_spend": 55000},
    {"id": 3, "name": "Sardor",  "monthly_spend": 180000},
    {"id": 4, "name": "Gulnora", "monthly_spend": 32000},
    {"id": 5, "name": "Timur",   "monthly_spend": 95000},
    {"id": 6, "name": "Madina",  "monthly_spend": 160000},
    {"id": 7, "name": "Otabek",  "monthly_spend": 47000},
    {"id": 8, "name": "Zarina",  "monthly_spend": 125000},
]

# Har bir daraja uchun hisoblagich — boshida hammasi 0
premium_count  = 0
standard_count = 0
basic_count    = 0


# Har bir mijoz bo'yicha sikl yuritamiz

for customer in customers_data:

    spend = customer["monthly_spend"]   # mijozning oylik xarajatini olamiz

    # Xarajatga qarab darajani aniqlaymiz
    if spend > 150000:
        tier = "premium"
        premium_count += 1              # premium hisoblagichni 1 ga oshiramiz

    elif spend >= 50000:
        tier = "standard"
        standard_count += 1             # standard hisoblagichni 1 ga oshiramiz

    else:
        tier = "basic"
        basic_count += 1                # basic hisoblagichni 1 ga oshiramiz

    # Har bir mijozning natijasini chiqaramiz
    print(f"  {customer['name']:<10} | Xarajat: {spend:>8,} so'm | Daraja: {tier}")


# YAKUNIY HISOBOT — sikldan tashqarida, hammasi tugagandan keyin

print("\n--- Xulosa ---")
print(f"Premium:  {premium_count} ta mijoz")
print(f"Standard: {standard_count} ta mijoz")
print(f"Basic:    {basic_count} ta mijoz")
print(f"Jami:     {premium_count + standard_count + basic_count} ta mijoz")
```
