```python
import pandas as pd

# Faylni yuklash
df = pd.read_csv('https://raw.githubusercontent.com/leonism/sample-superstore/master/data/superstore.csv')

# Birinchi 10 ta qatorni ko'rish
print(df.head(10))
```

```python
# Hajmi (qator va ustunlar soni)
print("Hajmi:", df.shape)

# Ma'lumot turlari va to'ldirilgan qatorlar haqida ma'lumot
df.info()

# Raqamli ustunlar uchun statistik xulosa
print(df.describe())

# Har bir ustundagi bo'sh (null) qiymatlar soni
print(df.isnull().sum())
```

```python
# Savdo (Sales) ustunidagi bo'shliqlarni mediana bilan to'ldirish
df['Sales'] = df['Sales'].fillna(df['Sales'].median())

# Yetkazib berish usuli (Ship Mode) ustunidagi bo'shliqlarni eng ko'p takrorlangan qiymat (moda) bilan to'ldirish
df['Ship Mode'] = df['Ship Mode'].fillna(df['Ship Mode'].mode()[0])

# Tekshirish: endi bo'sh joylar qolmaganini ko'ramiz
print(df.isnull().sum())
```

```python
# Takrorlangan qatorlar sonini aniqlash
print("Dublikatlar soni:", df.duplicated().sum())

# Dublikatlarni o'chirish va natijani saqlash
df = df.drop_duplicates()

# Yangi hajmni tekshirish
print("Tozalangandan keyingi hajm:", df.shape)
```

```python
# Matn ko'rinishidagi sanalarni maxsus datetime formatiga o'tkazish
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Tekshirish: ustunlar turi datetime64 ekanligiga ishonch hosil qilish
print(df.dtypes[['Order Date', 'Ship Date']])
```

```python
# 1. Har bir mijozning umumiy xarajatlari
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']

# 2. Har bir mijozning buyurtmalar soni
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']

# 3. Har bir mijozning o'rtacha buyurtma qiymati
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']

# Barcha ko'rsatkichlarni bitta jadvalga birlashtirish
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')

# Yakuniy natijalarni CSV fayllarga saqlash
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)

print("Tayyor! Fayllar saqlandi.")
```
