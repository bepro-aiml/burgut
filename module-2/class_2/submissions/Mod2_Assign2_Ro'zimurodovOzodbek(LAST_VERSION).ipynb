```python
# -----------------************** Task 3: Write a Function **************-----------------
# â¢ Step 1: In a new code cell, define the function:
def calculate_average(numbers):
    total = sum(numbers)
    count = len(numbers)
    return total / count
# â¢ Step 2: Test it with two different lists:
test1 = [80, 90, 70, 85, 95]
test2 = [55, 62, 48, 71]
print("Average 1:", calculate_average(test1))
print("Average 2:", calculate_average(test2))
# â¢ Step 3: Verify the output makes sense (mentally check one of the averages).
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("vivek468/superstore-dataset-final")

print("Path to dataset files:", path)
```

```python
# -----Task 2: Inspect the Structure-------
import pandas as pd

# Load the dataset from a CSV file into a Pandas DataFrame
df = pd.read_csv('/kaggle/input/superstore-dataset-final/Sample - Superstore.csv', encoding='latin1')

# Display the first 10 rows to verify the data was loaded correctly
df.head(10)
```

```python
# -----Task 3: Handle Missing Values-------
# Fill missing numerical values with the median
df['Sales'] = df['Sales'].fillna(df['Sales'].median())

# Fill missing categorical (text) values with the mode (most common value)
df['Ship Mode'] = df['Ship Mode'].fillna(df['Ship Mode'].mode()[0])

# Verify that there are no more missing values
print(df.isnull().sum())
```

```python
# ------Task 4: Remove Duplicates ------
# Check the number of duplicate rows
print(f"Duplicates found: {df.duplicated().sum()}")

# Remove duplicate rows
df = df.drop_duplicates()

# Verify the new shape of the dataset
print(f"New shape after removing duplicates: {df.shape}")
```

```python
# -----Task 5: Convert Data Types-------
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
print(df.dtypes)
df[['Order Date', 'Ship Date']].head()
```

```python
# -----Task 6: Create Derived Features-------
# Step 1: Calculate total spending per customer
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spent']

# Step 2: Calculate order frequency (how many times they shopped)
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_count']

# Step 3: Merge these into a customer summary table
customer_summary = total_spending.merge(order_freq, on='Customer ID')

# Step 4: Show the first 10 rows of the summary
customer_summary.head(10)
```

```python
# Save the cleaned data to a new CSV file
df.to_csv('superstore_cleaned.csv', index=False)

# Save the customer summary to a new CSV file
customer_summary.to_csv('customer_summary.csv', index=False)

print("Files saved successfully!")
```
