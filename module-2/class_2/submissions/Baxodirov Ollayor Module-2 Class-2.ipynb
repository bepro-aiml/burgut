## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# Count missing values per column
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]
```

```python
# Fill missing numerical values with median (robust to outliers)
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify: no more missing values
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")

# =========================

# Purpose:
# - Identify columns with missing (null) values
# - Fill numeric columns using median
# - Fill categorical columns using mode (most frequent value)

# Why it is needed:
# - Missing values can break analysis or models
# - Median is robust to outliers (better than mean)
# - Mode is appropriate for categorical data

# =========================

# EN: Count missing values per column
# UZ: Har bir ustundagi bo‘sh qiymatlarni sanaydi
# missing = df.isnull().sum()

# EN: Calculate missing percentage
# UZ: Bo‘sh qiymatlar foizi
# missing_pct = (missing / len(df) * 100).round(2)

# EN: Combine into report
# UZ: Hisobot jadvali yaratish
# missing_report = pd.DataFrame({
    # 'missing_count': missing,
    # 'missing_pct': missing_pct
# })

# EN: Show only columns with missing values
# UZ: Faqat bo‘sh qiymat bor ustunlar
# print(missing_report[missing_report['missing_count'] > 0])


# EN: Fill numeric columns with median
# UZ: Numeric ustunlarni median bilan to‘ldirish
# numerical_cols = df.select_dtypes(include=[np.number]).columns
# for col in numerical_cols:
    # if df[col].isnull().sum() > 0:
        # median_val = df[col].median()
        # df[col].fillna(median_val, inplace=True)


# EN: Fill categorical columns with mode
# UZ: Kategorik ustunlarni mode bilan to‘ldirish
# categorical_cols = df.select_dtypes(include=['object']).columns
# for col in categorical_cols:
    # if df[col].isnull().sum() > 0:
        # mode_val = df[col].mode()[0]
        # df[col].fillna(mode_val, inplace=True)


# EN: Verify no missing values remain
# UZ: Bo‘sh qiymatlar qolmaganini tekshirish
# print("Remaining missing:", df.isnull().sum().sum())
```
