```python
#Function
def calculate_student_average(grades):
    total = sum(grades)      # Adds all the scores together
    count = len(grades)     # Counts how many subjects there are
    return total / count     # Returns the average (as a float)

# Testing the function with two different students
student_a_grades = [85, 90, 70, 85, 95]
student_b_grades = [55, 62, 48, 71]

print("Student A Average:", calculate_student_average(student_a_grades))
print("Student B Average:", calculate_student_average(student_b_grades))
```

```python
#Putting all togerher
import random

# We need the function we talked about earlier
def calculate_average(numbers):
    return sum(numbers) / len(numbers)

# Step 1: Create a list of 10 random scores
exam_scores = [random.randint(0, 100) for _ in range(10)]
print("Scores:", exam_scores)

# Step 2: Use the function
avg = calculate_average(exam_scores)
print(f"Class average: {avg:.1f}")

# Step 3: Add a pass/fail check (Threshold: 60)
if avg >= 60:
    print("Result: PASS")
else:
    print("Result: FAIL")
```
