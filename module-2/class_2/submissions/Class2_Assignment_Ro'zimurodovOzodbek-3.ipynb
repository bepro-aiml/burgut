## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# -----------------************** FIFTH CELL **************-----------------
# Count missing values per column
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]
```

```python
# FIFTH CELL explanation:
# 1. We count how many values are missing in each column using df.isnull().sum()
# 2. We also calculate the percentage of missing values per column
# 3. The report shows that 806 rows (7.46%) have missing values in most columns
# 4. Postal Code has 817 missing values - slightly more than the others
```

```python
# -----------------************** SIXTH CELL **************-----------------
# Fill missing numerical values with median (robust to outliers)
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify: no more missing values
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")
```

```python
# SIXTH CELL explanation:
# 1. For numerical columns (Sales, Profit, etc.) we fill missing values with the median
# 2. We use median instead of mean because it is not affected by extreme outliers
# 3. For categorical columns (Ship Mode, City, etc.) we fill with the mode (most frequent value)
# 4. After filling, Total missing values remaining: 0 - all missing data is handled
```

## Task 3: Remove Duplicates (pre-built)

```python
# -----------------************** SEVENTH CELL **************-----------------
# Check for duplicates
n_duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {n_duplicates}")

# Remove duplicates
if n_duplicates > 0:
    df = df.drop_duplicates()
    print(f"After removal: {df.shape[0]} rows remain")
else:
    print("No duplicates to remove.")
```

```python
# SEVENTH CELL explanation:
# 1. We check for duplicate rows using df.duplicated().sum()
# 2. The result shows 0 duplicate rows found in this dataset
# 3. No rows needed to be removed, so the dataset stays at its original size
```
