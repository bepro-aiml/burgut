```python
# Ro'yxatdagi sonlarning o'rtacha qiymatini hisoblash funksiyasi
def calculate_average(numbers):

    # Barcha sonlar yig'indisini topish
    total = sum(numbers)

    # Elementlar sonini topish
    count = len(numbers)

    # O'rtacha qiymatni qaytarish
    return total / count
```

```python
# 1-test natijalari ro'yxati
test1 = [80, 90, 70, 85, 95]

# 2-test natijalari ro'yxati
test2 = [55, 62, 48, 71]

# 1-test o'rtacha bahosini chiqarish
print("Average 1:", calculate_average(test1))

# 2-test o'rtacha bahosini chiqarish
print("Average 2:", calculate_average(test2))
```

```python
# Shaharlar nomlari saqlangan ro'yxat
cities = ["Tashkent", "Samarkand", "Bukhara", "Namangan", "Fergana"]

# Har bir shaharni index raqami bilan chiqarish
for index, city in enumerate(cities):

    # Index va shahar nomini ekranga chiqarish
    print(f"Index {index}: {city}")
```

```python
# Random sonlar yaratish uchun kutubxona chaqirish
import random

# 10 ta tasodifiy imtihon ballari ro'yxatini yaratish
exam_scores = [random.randint(0, 100) for _ in range(10)]

# Ballarni ekranga chiqarish
print("Scores:", exam_scores)

# O'rtacha ballni hisoblash
avg = calculate_average(exam_scores)

# O'rtacha natijani 1 xonali kasr bilan chiqarish
print(f"Class average: {avg:.1f}")

# O'rtacha ball 60 yoki undan yuqori bo'lsa PASS
if avg >= 60:
    print("Result: PASS")

# Aks holda FAIL
else:
    print("Result: FAIL")
```
