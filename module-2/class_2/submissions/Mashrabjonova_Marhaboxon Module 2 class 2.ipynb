## Task 2: Missing Values (pre-built)

Check which columns have missing values and how many.

```python
# Count missing values per column
missing = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_report = pd.DataFrame({
    'missing_count': missing,
    'missing_pct': missing_pct
})

# Show only columns with missing values
missing_report[missing_report['missing_count'] > 0]

# Calculates: It counts how many empty cells (nulls) exist in each column and calculates their percentage.

# Summarizes: It combines these two numbers into a new table called missing_report.

# Filters: It only displays columns that actually have missing values, hiding the ones that are "clean."
```

```python
# Fill missing numerical values with median (robust to outliers)
numerical_cols = df.select_dtypes(include=[np.number]).columns
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        median_val = df[col].median()
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing values with median: {median_val}")

# Fill missing categorical values with mode
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        mode_val = df[col].mode()[0]
        df[col].fillna(mode_val, inplace=True)
        print(f"Filled {col} missing values with mode: {mode_val}")

# Verify: no more missing values
print(f"\nTotal missing values remaining: {df.isnull().sum().sum()}")

# Numerical Data: It finds columns with numbers and fills empty cells with the median. The median is used because it isn't skewed by extreme "outlier" values.

# Categorical Data: It finds columns with text/labels and fills empty cells with the mode (the most frequently occurring value).

# Verification: It prints a final confirmation showing that the total number of missing values is now 0.
```

## Task 3: Remove Duplicates (pre-built)

```python
# Check for duplicates
n_duplicates = df.duplicated().sum()
print(f"Duplicate rows found: {n_duplicates}")

# Remove duplicates
if n_duplicates > 0:
    df = df.drop_duplicates()
    print(f"After removal: {df.shape[0]} rows remain")
else:
    print("No duplicates to remove.")

#     Detection: It uses df.duplicated().sum() to count how many rows are exact copies of others.

# Conditional Removal: If duplicates exist (count > 0), it deletes the extra copies using df.drop_duplicates().

# Verification: It prints the final row count to show you how much the data "shrunk" after the cleanup.
```

Import and Load the Data

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('SampleSuperstore.csv')

# Verify by showing the first 10 rows
df.head(10)
```

Task 2: Inspect the Data Structure

```python
# Check shape, info and basic statistics
print("Dataset Shape:", df.shape)
df.info()
print(df.describe())

# Check for missing values
print("Missing Values:\n", df.isnull().sum())
```

Task 3: Handle Missing Values

```python
# Fill missing values for numerical columns with median
df['Sales'] = df['Sales'].fillna(df['Sales'].median())

# Fill missing values for categorical columns with mode
df['Ship Mode'] = df['Ship Mode'].fillna(df['Ship Mode'].mode()[0])

print("Missing values after cleaning:", df.isnull().sum().sum())
```

Task 4 & 5: Remove Duplicates and Convert Date Formats

```python
# Remove duplicate rows
df = df.drop_duplicates()

# Convert Date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

print("Duplicates removed and date formats converted.")
```

Task 6: Customer Analytics and Saving

```python
# Create customer-level metrics
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()

# Merge into a summary table
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')
customer_summary.columns = ['Customer ID', 'total_spending', 'order_frequency', 'avg_order_value']

# Save the results
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)

customer_summary.head(10)
```
