````python
import pandas as pd

file_name = 'Sample - Superstore.csv'
df = pd.read_csv(file_name, encoding='latin1')

display(df.head())
#faylni o'qiymiz
````

````python
print(df.isnull().sum())
# ma'lumotlar orasida bo'sh joy borligini tekshiramiz
````

Xo'sh, datasetda 9994 ta qator bor, .isnull() natijasiga ko'ra bo'sh joy yo'q. Do'kon ba'zi savdolarda juda katta zarar ko'ryapti. Ba'zi mahsulotlarga 80 foizgacha chegirma bo'lgan.

````python
df['Sales'].fillna(df['Sales'].median(), inplace=True)
df['Ship Mode'].fillna(df['Ship Mode'].mode()[0], inplace=True)
print(df.isnull().sum())

#bo'sh joy yo'q edi, shu sabab tashlab o'tsak ham zarar qilmasdi task 3 da aytilgan topshiriqlarni
````

Bo'sh qiymatlar yoq!

````python
# --- TASK 4: Duplikattlarni o'chirish ---
print("Duplicate rows before:", df.duplicated().sum())
df = df.drop_duplicates()
print("Shape after removing duplicates:", df.shape)

# --- TASK 5: Sana formatiga o'tkazish ---
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

# Natijani tekshirish
print("\nMa'lumot turlari:")
print(df.dtypes[['Order Date', 'Ship Date']])
df[['Order Date', 'Ship Date']].head()
````

Datasetda duplikatlar tekshirildi va o'chirildi. Sana ustunlari datetime formatiga o'tkazildi

````python
# 1. Har bir mijoz uchun jami xarajat
total_spending = df.groupby('Customer ID')['Sales'].sum().reset_index()
total_spending.columns = ['Customer ID', 'total_spending']

# 2. Buyurtmalar chastotasi (necha marta buyurtma bergan)
order_freq = df.groupby('Customer ID')['Order ID'].nunique().reset_index()
order_freq.columns = ['Customer ID', 'order_frequency']

# 3. O'rtacha buyurtma qiymati
avg_order = df.groupby('Customer ID')['Sales'].mean().reset_index()
avg_order.columns = ['Customer ID', 'avg_order_value']

# 4. Hamma ma'lumotlarni bitta mijozlar jadvaliga birlashtirish
customer_summary = total_spending.merge(order_freq, on='Customer ID').merge(avg_order, on='Customer ID')

# 5. Tozalangan va yangi jadvallarni CSV sifatida saqlash
df.to_csv('superstore_cleaned.csv', index=False)
customer_summary.to_csv('customer_summary.csv', index=False)

# Natijani ko'rish
print("Mijozlar statistikasi tayyor!")
customer_summary.head(10)
````
